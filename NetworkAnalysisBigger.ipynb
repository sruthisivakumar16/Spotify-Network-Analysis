```python
import random
import matplotlib.pyplot as plt
import networkx as nx

# Assume bigger_graph is your NetworkX graph and bigger_closeness_values contains closeness centrality values
threshold = np.percentile(bigger_closeness_values, 90)  # For example, select top 10%

# Identify nodes with closeness centrality above the threshold
top_nodes = [node for node, closeness in nx.closeness_centrality(bigger_graph).items() if closeness >= threshold]

# Randomly select a subset of nodes from the top nodes
num_random_nodes = 30  # Number of random nodes to select
random_core_nodes = random.sample(top_nodes, min(num_random_nodes, len(top_nodes)))

# Create the subgraph
subgraph = bigger_graph.subgraph(random_core_nodes)

# Get song names from the song_data DataFrame
song_names = {node: song_data[song_data['song_id'] == node]['song_name'].values[0] for node in subgraph.nodes}

# Get node centrality values for color mapping
node_centrality_values = {node: nx.closeness_centrality(bigger_graph)[node] for node in subgraph.nodes}
node_colors = [node_centrality_values[node] for node in subgraph.nodes]

# Plot the subgraph with song names as labels and node color according to centrality values
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(subgraph)  # Position nodes using a spring layout
nx.draw(subgraph, pos, labels=song_names, with_labels=True, node_color=node_colors, cmap=plt.cm.viridis, node_size=500, font_size=10)
plt.title('Subgraph of Closeness Centrality of Random Nodes')
plt.colorbar(label='Closeness Centrality')  # Add color bar indicating the centrality values
plt.show()
```

```python
#Calculate Freeman Centralisation score based on the formula
def freeman_centralisation(G):
    dc = [G.degree(node) for node in G.nodes()]
    max_dc = max(dc)
    n = len(dc)
    numerator = sum([max_dc - c for c in dc])
    denominator = (n-1)*(n-2)
    return float(numerator)/denominator
```
